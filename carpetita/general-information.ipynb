```python
# Extraer todos todas las ofertas por categoría y muestra cuantas ofertas hay por categoría, el total, 
# y la categoría con más ofertas y la categoría con menos ofertas

import pandas as pd

# Cargar el archivo CSV en un DataFrame
df = pd.read_csv('../data/all_jobs_data.csv')

# Contar el número de ofertas por categoría
job_counts = df['category'].value_counts()

# Categoría con más ofertas de trabajo
most_jobs_category = job_counts.idxmax()
most_jobs_count = job_counts.max()

# Categoría con menos ofertas de trabajo
least_jobs_category = job_counts.idxmin()
least_jobs_count = job_counts.min()

# Mostrar el número de ofertas por categoría
print(job_counts)

# Mostrar la categoría con más ofertas de trabajo
print(f"\nLa categoría con más ofertas de trabajo es '{most_jobs_category}' con {most_jobs_count} ofertas.")

# Mostrar la categoría con menos ofertas de trabajo
print(f"La categoría con menos ofertas de trabajo es '{least_jobs_category}' con {least_jobs_count} ofertas.")

# Calcular y mostrar el total de ofertas de trabajo
total_job_offers = job_counts.sum()
print(f"\nTotal de ofertas de trabajo: {total_job_offers}")
```

```python
#  Calcular el salario promedio, mínimo y máximo por categoría y mostrarlo

import pandas as pd

# Cargar el archivo CSV en un DataFrame
df = pd.read_csv('../data/all_jobs_data.csv')

# Eliminar filas con valores nulos en 'min_salary' y 'max_salary'
df = df.dropna(subset=['min_salary', 'max_salary'])

# Calcular el salario promedio
df['average_salary'] = (df['min_salary'] + df['max_salary']) / 2

# Agrupar por categoría y calcular el salario promedio, mínimo y máximo
salary_stats_by_category = df.groupby('category').agg({
    'min_salary': 'mean',
    'max_salary': 'mean',
    'average_salary': 'mean'
}).reset_index()

# Renombrar las columnas para mayor claridad
salary_stats_by_category.columns = ['category', 'average_min_salary', 'average_max_salary', 'average_salary']

# Mostrar los resultados
print(salary_stats_by_category.to_string(index=False))

```

```python
# Calcular el salario promedio por categoría y mostrar la categoría con el salario promedio más alto

import pandas as pd

df = pd.read_csv('../data/all_jobs_data.csv')

df = df.dropna(subset=['min_salary', 'max_salary'])

df['average_salary'] = (df['min_salary'] + df['max_salary']) / 2

average_salary_by_category = df.groupby('category')['average_salary'].mean()

highest_avg_salary_category = average_salary_by_category.idxmax()
highest_avg_salary = average_salary_by_category.max()

print(f"La categoría con el promedio de pago más alto es '{highest_avg_salary_category}' con un salario promedio de {highest_avg_salary}.")
```
